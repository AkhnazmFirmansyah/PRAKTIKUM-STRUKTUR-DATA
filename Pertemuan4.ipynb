**Tugas/Latihan 1**

```python
class TicketQueue:
    def __init__(self, capacity=10):
        self.queue = []
        self.capacity = capacity

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def enqueue(self, ticket_id):
        # Cek apakah pelanggan adalah VIP (mengandung angka 7)
        if '7' in ticket_id:
            print(f"VIP Customer {ticket_id} langsung mendapat tiket tanpa antri.")
        else:
            if self.size() < self.capacity:
                self.queue.append(ticket_id)
                print(f"Pelanggan {ticket_id} masuk ke antrian.")
            else:
                print(f"Antrian penuh! Kapasitas maksimal {self.capacity}. Tunggu hingga ada slot kosong.")

    def dequeue(self):
        if not self.is_empty():
            served = self.queue.pop(0)
            print(f"Pelanggan {served} telah dilayani.")
        else:
            print("Antrian kosong! Tidak ada pelanggan yang bisa dilayani.")

    def front(self):
        return self.queue[0] if not self.is_empty() else "Antrian kosong"

    def rear(self):
        return self.queue[-1] if not self.is_empty() else "Antrian kosong"

    def display(self):
        if not self.is_empty():
            print("Antrian:", " -> ".join(self.queue))
        else:
            print("Antrian kosong.")

# Program utama dengan input pengguna
queue = TicketQueue()

while True:
    print("\nMenu Antrian Tiket Bioskop:")
    print("1. Tambah pelanggan ke antrian")
    print("2. Layani pelanggan")
    print("3. Tampilkan antrian")
    print("4. Tampilkan pelanggan pertama dan terakhir")
    print("5. Keluar")

    pilihan = input("Pilih menu (1-5): ")

    if pilihan == "1":
        ticket_id = input("Masukkan kode pelanggan (2 huruf nama + 2 digit terakhir NPM): ").strip().upper()
        queue.enqueue(ticket_id)
    elif pilihan == "2":
        queue.dequeue()
    elif pilihan == "3":
        queue.display()
    elif pilihan == "4":
        print("Front:", queue.front())
        print("Rear:", queue.rear())
    elif pilihan == "5":
        print("Keluar dari sistem antrian tiket.")
        break
    else:
        print("Pilihan tidak valid! Silakan coba lagi.")
```

**Tugas/Latihan 2**

```python
class PhotocopyQueue:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def enqueue(self, order_id, num_pages):
        # Fast Track: Jika halaman kurang dari 10, langsung diproses
        if num_pages < 10:
            print(f"Pesanan {order_id} langsung diproses tanpa harus mengantri (Fast Track).")
        else:
            self.queue.append((order_id, num_pages))
            print(f"Pesanan {order_id} ({num_pages} halaman) masuk ke antrian.")

    def dequeue(self):
        if not self.is_empty():
            served_order = self.queue.pop(0)
            order_id, num_pages = served_order
            print(f"Pesanan {order_id} ({num_pages} halaman) telah dilayani.")
            # Jika halaman lebih dari 100, proses dalam dua tahap
            if num_pages > 100:
                print(f"Pesanan {order_id} diproses dalam dua tahap.")
        else:
            print("Antrian kosong! Tidak ada pesanan yang bisa dilayani.")

    def front(self):
        return self.queue[0] if not self.is_empty() else "Antrian kosong"

    def rear(self):
        return self.queue[-1] if not self.is_empty() else "Antrian kosong"

    def display(self):
        if not self.is_empty():
            print("Antrian Fotokopi:")
            for order in self.queue:
                print(f"{order[0]} - {order[1]} halaman")
        else:
            print("Antrian kosong.")

# Program utama dengan input pengguna
queue = PhotocopyQueue()

while True:
    print("\nMenu Antrian Fotokopi Kampus:")
    print("1. Tambah pesanan fotokopi")
    print("2. Layani pesanan")
    print("3. Tampilkan antrian")
    print("4. Tampilkan pesanan pertama dan terakhir")
    print("5. Keluar")

    pilihan = input("Pilih menu (1-5): ")

    if pilihan == "1":
        order_id = input("Masukkan ID pesanan (3 huruf nama + 2 digit terakhir NPM): ").strip().upper()
        num_pages = int(input("Masukkan jumlah halaman: "))
        queue.enqueue(order_id, num_pages)
    elif pilihan == "2":
        queue.dequeue()
    elif pilihan == "3":
        queue.display()
    elif pilihan == "4":
        print("Front:", queue.front())
        print("Rear:", queue.rear())
    elif pilihan == "5":
        print("Keluar dari sistem antrian fotokopi.")
        break
    else:
        print("Pilihan tidak valid! Silakan coba lagi.")
```

**Tugas/Latihan 3**

```python
class ParkingQueue:
    def __init__(self, capacity=15):
        self.queue = []
        self.capacity = capacity  # Kapasitas maksimal parkiran

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def enqueue(self, vehicle_number):
        # Kendaraan listrik mendapatkan prioritas (mengandung 'E' untuk Electric)
        if 'E' in vehicle_number:
            self.queue.insert(0, vehicle_number)
            print(f"Parkir Prioritas: Kendaraan {vehicle_number} langsung masuk.")
        else:
            # Cek apakah parkiran penuh
            if len(self.queue) < self.capacity:
                self.queue.append(vehicle_number)
                print(f"Kendaraan {vehicle_number} masuk ke parkiran.")
            else:
                print(f"Parkiran penuh! Kendaraan {vehicle_number} harus menunggu.")

    def dequeue(self):
        if not self.is_empty():
            vehicle = self.queue.pop(0)
            print(f"Kendaraan {vehicle} keluar dari parkiran.")
        else:
            print("Parkiran kosong, tidak ada kendaraan yang keluar.")

    def front(self):
        if not self.is_empty():
            return self.queue[0]
        return "Parkiran kosong"

    def rear(self):
        if not self.is_empty():
            return self.queue[-1]
        return "Parkiran kosong"

    def display(self):
        if not self.is_empty():
            print("Kendaraan dalam parkiran:", " -> ".join(self.queue))
        else:
            print("Parkiran kosong.")

# Program utama dengan input pengguna
parking = ParkingQueue()

while True:
    print("\nMenu Parkiran:")
    print("1. Masukkan kendaraan")
    print("2. Keluarkan kendaraan")
    print("3. Tampilkan antrian parkir")
    print("4. Tampilkan kendaraan pertama dan terakhir")
    print("5. Keluar")

    pilihan = input("Pilih menu (1-5): ")

    if pilihan == "1":
        vehicle_number = input("Masukkan nomor kendaraan: ").strip().upper()
        parking.enqueue(vehicle_number)
    elif pilihan == "2":
        parking.dequeue()
    elif pilihan == "3":
        parking.display()
    elif pilihan == "4":
        print("Front:", parking.front())
        print("Rear:", parking.rear())
    elif pilihan == "5":
        print("Keluar dari sistem parkiran.")
        break
    else:
        print("Pilihan tidak valid! Silakan coba lagi.")
```

**Tugas/Latihan 4**

```python
class CertificateQueue:
    def __init__(self):
        self.regular_queue = []  # Antrian mahasiswa reguler
        self.achiever_queue = []  # Antrian mahasiswa berprestasi

    def is_empty(self):
        return len(self.regular_queue) == 0 and len(self.achiever_queue) == 0

    def size(self):
        return len(self.regular_queue) + len(self.achiever_queue)

    def enqueue(self, student_id, is_achiever=False):
        # Jika mahasiswa berprestasi, masuk ke antrian prioritas
        if is_achiever:
            self.achiever_queue.append(student_id)
            print(f"Mahasiswa berprestasi {student_id} langsung mendapat sertifikat (Prioritas).")
        else:
            self.regular_queue.append(student_id)
            print(f"Mahasiswa {student_id} masuk ke antrian reguler.")

        # Jika antrian melebihi 20, bagi menjadi 2 jalur
        if self.size() > 20:
            print("Antrian melebihi 20 orang, membagi menjadi jalur reguler dan jalur berprestasi.")

    def dequeue(self):
        if not self.is_empty():
            if self.achiever_queue:
                student = self.achiever_queue.pop(0)
                print(f"Mahasiswa berprestasi {student} mengambil sertifikat.")
            else:
                student = self.regular_queue.pop(0)
                print(f"Mahasiswa {student} mengambil sertifikat.")
        else:
            print("Tidak ada mahasiswa dalam antrian.")

    def front(self):
        if not self.is_empty():
            if self.achiever_queue:
                return self.achiever_queue[0]
            return self.regular_queue[0]
        return "Antrian kosong"

    def rear(self):
        if not self.is_empty():
            if self.regular_queue:
                return self.regular_queue[-1]
            return self.achiever_queue[-1]
        return "Antrian kosong"

    def display(self):
        if not self.is_empty():
            print("Antrian mahasiswa berprestasi:", " -> ".join(self.achiever_queue) if self.achiever_queue else "Kosong")
            print("Antrian mahasiswa reguler:", " -> ".join(self.regular_queue) if self.regular_queue else "Kosong")
        else:
            print("Antrian kosong.")

# Program utama dengan input pengguna
queue = CertificateQueue()

while True:
    print("\nMenu Antrian Sertifikat:")
    print("1. Masukkan mahasiswa ke antrian")
    print("2. Layani mahasiswa")
    print("3. Tampilkan antrian")
    print("4. Tampilkan mahasiswa pertama dan terakhir")
    print("5. Keluar")

    pilihan = input("Pilih menu (1-5): ")

    if pilihan == "1":
        student_id = input("Masukkan ID Mahasiswa (huruf program studi + 2 digit NPM): ").strip().upper()
        prestasi = input("Apakah mahasiswa berprestasi? (y/n): ").strip().lower() == "y"
        queue.enqueue(student_id, prestasi)
    elif pilihan == "2":
        queue.dequeue()
    elif pilihan == "3":
        queue.display()
    elif pilihan == "4":
        print("Front:", queue.front())
        print("Rear:", queue.rear())
    elif pilihan == "5":
        print("Keluar dari sistem antrian sertifikat.")
        break
    else:
        print("Pilihan tidak valid! Silakan coba lagi.")
```
