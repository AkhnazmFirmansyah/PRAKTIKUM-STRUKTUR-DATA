```python
# Ukuran peta
size = 10

# Membuat peta dengan harta karun tersembunyi
treasure_map = ["_" for _ in range (size)]

# Tentukan lokasi harta karun secara manual
print(f"Selamat datang di Treasure Hunt! Peta memiliki {size} lokasi (0 hingga {size-1}).")
treasure_index = int(input(f"Masukkan lokasi harta karun (0 hingga {size-1}): "))
while treasure_index < 0 or treasure_index >= size:
    print("Lokasi tidak valid. Masukkan angka dalam rentang yang benar.")
    treasure_index = int(input(f"Masukkan lokasi harta karun (0 hingga {size-1}): "))

treasure_map[treasure_index] = "X" # Menyembunyikan harta karun di lokasi yang dipilih

# Fungsi untuk memulai permainan
def play_treasure_hunt ():
    print("\nHarta karun telah disembunyikan! Sekarang giliran pemain menebak.")
    print("Cobalah untuk menemukan harta karun dengan menebak indeks yang benar.")

    # Loop permainan
    while True:
        # Menampilkan peta tersembunyi kepada pemain
        print("\nPeta saat ini: ", ["_" for _ in range (size)]) # Peta tersembunyi

        # Input pemain
        try:
            guess = int(input(f"Masukkan indeks tebakanmu (0-{size-1}):"))
        except ValueError:
            print("Masukkan angka yang valid.")
            continue

        # Validasi input
        if guess < 0 or guess >= size:
            print("Indeks tidak valid, pilih antara 0 hingga", size-1)
            continue

        # Mengecek apakah tebakan benar
        if treasure_map[guess] == "X":
           print("🎉 Selamat! kamu menemukan harta karun di lokasi:", guess)
           break
        else:
           print("❌ Harta karun tidak ditemukan di lokasi ini. Coba lagi!")

    print("Terima kasih sudah bermain!")

# Jalankan permainan
play_treasure_hunt()
```

**Sesudah Dimodifikasi**

```python
import random

# Fungsi untuk memulai permainan
def play_treasure_hunt():
    # Meminta input dari pengguna untuk menentukan ukuran peta
    while True:
        try:
            size = int(input("Masukkan ukuran peta (misalnya 10): "))
            if size <= 1:
                print("Ukuran peta harus lebih besar dari 1.")
            else:
                break
        except ValueError:
            print("Masukkan angka yang valid untuk ukuran peta.")

    # Membuat peta dengan ukuran yang ditentukan
    treasure_map = ["_" for _ in range(size)]

    # Mengacak lokasi harta karun
    treasure_location = random.randint(0, size - 1)
    treasure_map[treasure_location] = "X"  # Menandai lokasi harta karun

    # Menyimpan tebakan pemain untuk menandai tebakan sebelumnya
    guesses = []

    print("\nHarta karun telah disembunyikan! Sekarang giliran pemain menebak.")
    print("Cobalah untuk menemukan harta karun dengan menebak indeks yang benar.")

    # Loop permainan
    while True:
        # Menampilkan peta tersembunyi dengan tanda pada tebakan sebelumnya
        current_map = ["X" if i in guesses and i != treasure_location else "_" for i in range(size)]
        print("\nPeta saat ini: ", current_map)

        # Input pemain
        try:
            guess = int(input("Masukkan indeks tebakanmu : "))
        except ValueError:
            print("Masukkan angka yang valid.")
            continue

        # Validasi input
        if guess < 0 or guess >= size:
            print("Indeks tidak valid, pilih antara 0 hingga", size - 1)
            continue

        # Mengecek apakah tebakan benar
        if guess == treasure_location:
            print(f"🎉 Selamat! Kamu menemukan harta karun di lokasi: {guess}")
            break
        else:
            if guess not in guesses:
                guesses.append(guess)  # Menambahkan tebakan ke daftar tebakan sebelumnya
            print("❌ Harta karun tidak ditemukan di lokasi ini. Coba lagi!")

    print("Terima kasih sudah bermain!")

# Jalankan permainan
play_treasure_hunt()
```
